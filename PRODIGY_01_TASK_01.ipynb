```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample dataset
data = {
    'Age': [23, 45, 31, 35, 26, 67, 34, 54, 29, 37, 41, 58, 22, 33, 40],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']
}

df = pd.DataFrame(data)

# Save the dataset to a CSV file
df.to_csv('D:/sample_population.csv', index=False)

# Load the data from the specified path
file_path = 'D:/sample_population.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print(df.head())

# Plot the histogram for 'Age'
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plot the bar chart for 'Gender'
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(10, 6))
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Distribution of Genders')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
